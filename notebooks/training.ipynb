```python
import pandas as pd
data=pd.read_csv('../data/processed/motion.csv')
data.head()
data.drop(['Gesture','Image Path'],axis=1,inplace=True)
```

```python
#train test split 
from sklearn.model_selection import train_test_split
import numpy as np
train_x,test_x,train_y,test_y = train_test_split(data,np.ones((data.shape[0])),test_size=0.2,random_state=42)
```

```python
import tensorflow as tf 

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(42,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax') 
])

model.compile(optimizer='adam',
             loss='binary_crossentropy', 
             metrics=['accuracy'])

model.fit(train_x,train_y, epochs=100)
```

```python
predictions=model.predict(test_x)
print(predictions)
model.save('../models/NN.h5')
```
